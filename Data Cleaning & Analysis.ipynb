# Dropping Irrelevant columns

```python
#1 I will have to drop all columns relating to collection, since this analysis is only focus on track(a single music)

#List of columns to drop
columns_to_drop = ['collectionName', 'collectionId', 'collectionArtistId', 'collectionArtistName', 'collectionViewUrl', 
                   'collectionCensoredName', 'collectionPrice', 'collectionExplicitness', 'collectionArtistViewUrl']
df = df.drop(columns=columns_to_drop, axis = 1)
```

```python
df
```

```python
#2 I  will have to drop columns that have repetitive records througout, but first of all let me locate such columns

unique_values = df.apply(lambda x: pd.unique(x))
unique_values
```

```python
#So from the above it is reasonable to drop column wrapperType and kind
columns_to_drop_1 = ['wrapperType', 'kind']
df = df.drop(columns=columns_to_drop_1, axis = 1)
```

```python
df
```

```python
#3 Now let us check the uniqueness between trackName and trackCensoredName

pd.unique(df['trackName'])
```

```python
pd.unique(df['trackCensoredName'])
```

```python
#As you can above both columns have the same values, but instead of assuming we can do further confirmation

are_columns_equal = (df['trackName'] == df['trackCensoredName']).all()
are_columns_equal
```

```python
#Our assumption was actually wrong there is a difference between both columns, but since our analysis will have nothing to do
# with trackCensoredName we will have to drop it

df = df.drop(['trackCensoredName'], axis = 1)
df.head()
```

```python
#4 We will also drop all columns relating to url since it will have no relevance in our analysis

columns_to_drop_2 = ['artistViewUrl', 'trackViewUrl', 'previewUrl', 'artworkUrl30', 'artworkUrl60', 'artworkUrl100']
df = df.drop(columns=columns_to_drop_2, axis = 1)
df.head()
```

```python
#5 We will also remove other columns that are not also relevant to analysis

columns_to_drop_3 = ['discCount', 'discNumber', 'trackCount', 'trackNumber', 'trackTimeMillis']
df = df.drop(columns=columns_to_drop_3, axis = 1)
df.head()
```

```python
#6 There are two colummns i will also have to drop again, the reason i'm dropping them is because since our analysis is based
# on the US there is no of having a column for that, and also the currency

columns_to_drop_4 = ['country', 'currency']
df = df.drop(columns=columns_to_drop_4, axis = 1)
df.head()
```

```python
#7 Finally we have to drop the artistId and trackId, because it is not also important for our analysis too

columns_to_drop_5 = ['artistId', 'trackId']
df = df.drop(columns=columns_to_drop_5, axis = 1)
df.head()
```

# Handling Missing Values

```python
# Now let us check if they are missing values in the dataframe we have now
missing_values = df.isnull().sum()
missing_values
```

```python
# Let us deal with the trackPrice, we can just the mean to fill out all missing values

col_mean = df['trackPrice'].mean()
df['trackPrice'] = df['trackPrice'].fillna(col_mean)
```

```python
df
```

# Checkpoint

```python
df_mod = df.copy()
df_mod
```

```python
# Dropping the only one row with an nan in the releaseDate column
df_mod = df_mod.dropna(subset=['releaseDate'])
```

```python
#convert the releaseDate to a datetime data type instead of an object, this is done for us to get rid of the time include with
# the date and work with the date only
df_mod['releaseDate'] = pd.to_datetime(df_mod['releaseDate'])
```

```python
df_mod['releaseDate']
```

```python
# We will have to get rid of the time 
df_mod.loc[:, 'releaseDate'] = df_mod['releaseDate'].dt.date
```

```python
df_mod['releaseDate']
```

```python
# We will have to convert the releaseDate back to datetime because after getting rid of the time it was convert to an object
df_mod['releaseDate'] = pd.to_datetime(df_mod['releaseDate'])
```

```python
df_mod['releaseDate']
```

```python
# Converting the releaseDate from this format yyyy-mm-dd to d/m/yyyy
df_mod['releaseDate'] = df_mod['releaseDate'].dt.strftime('%d/%m/%Y')
```

```python
df_mod['releaseDate']
```

```python
# I will have to drop the contentAdvisoryRating because our total rows is 10622 and the column is having a missng value 
# of 9099 which is almost 10000. So i will have to also drop it because it will be of no help for our analysis

df_mod = df_mod.dropna(axis=1, how='any')
```

```python
df_mod
```

# Handling The Negative Values in Track Price Column

```python
# Checking for the negative values in the trackPrice column
pd.unique(df_mod['trackPrice'])
```

```python
# Checking out how many they are in the column
negative_values_on_track_price = (df_mod['trackPrice'] < 0).sum()
negative_values_on_track_price
```

```python
# Dropping them because the number will not have that much impact in analysis
df_mod = df_mod.loc[df_mod['trackPrice'] >= 0].reset_index(drop=True)
```

```python
df_mod
```

```python
# Checking for the negative numbers to make everything is cleared
pd.unique(df_mod['trackPrice'])
```

```python
value_mapping = {
    'cleaned': 'Cleaned',
    'explicit': 'Explicit',
    'notExplicit': 'Not Explicit'
}

df_mod['trackExplicitness'] = df_mod['trackExplicitness'].replace(to_replace=value_mapping)
```
